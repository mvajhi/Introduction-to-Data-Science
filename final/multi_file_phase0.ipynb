```python
# %%
import pandas as pd
from jdatetime import datetime, date
from IPython.display import display

# %%
def run(path)->pd.DataFrame:
    df = pd.read_excel(path)

    df['ReserveStatus'].unique()
    
    # %%
    def convert_ReserveStatus(x):
        if(x == 'دریافت نشده' or x == 'ارسال نشده' or x == 'منقضی شده'):
            return 0
        return 1
    
    # %%
    df['ReserveStatus'] = df['ReserveStatus'].apply(convert_ReserveStatus)
    
    # %%
    df
    
    # %%
    df['ReserveStatus'].unique()
    
    # %%
    df['Reception'].unique()
    
    # %%
    def convert_Reception(x):
        if(x == 'کد فراموشی' or x == 'بارکد فراموشی'):
            return 1
        return 0
    
    # %%
    df['Reception'] = df['Reception'].apply(convert_Reception)
    
    # %%
    df['Reception'].unique()
    
    # %%
    def apply_recive_with_card(row):
        if row['Reception'] == 0 and row['ReserveStatus'] == 1:
            return 1
        return 0
    
    # %%
    df['ReceiveWithCard'] = df.apply(apply_recive_with_card,axis=1)
    
    # %%
    def apply_reciev_with_code(row):
        if row['Reception'] == 1 and row['ReserveStatus'] == 1:
            return 1
        return 0
    
    # %%
    df['ReceiveWithCode'] = df.apply(apply_reciev_with_code,axis=1)
    
    # %%
    def apply_dont_reciev(row):
        if row['ReserveStatus'] == 0:
            return 1
        return 0
    
    # %%
    df['DontReceive'] = df.apply(apply_dont_reciev,axis=1)
    
    # %%
    df.info()
    
    # %%
    df['DateReserve'] = df['DateReserve'].astype(str)
    
    # %%
    df.info()
    
    # %%
    df
    
    # %%
    result = df
    result
    
    # %%
    result = result[result['RestaurantName'].str.contains('رستوران آزاد') == False]
    result = result[result['RestaurantName'].str.contains('مهمان') == False]
    
    
    # %%
    result = result.drop(result[result['Name'] == 'سحر'].index)
    result = result.drop(result[result['Name'] == 'صبحانه'].index)
    
    # %%
    result = result.reset_index(drop=True)
    
    # %%
    result['Name'].unique()
    
    # %%
    def delete_definition(x):
        sep = '+'
        stripped = x.split(sep, 1)[0]
        return stripped
    result['FoodName'] = result['FoodName'].apply(delete_definition)
    
    # %%
    def delete_gender_from_name(x:str):
        out = x
        for i in ['خواهران', 'برادران', 'خانم', 'آقا', 'دختران', 'پسران']:
            out = out.replace(i, '')
        return out
    result['RestaurantName'] = result['RestaurantName'].apply(delete_gender_from_name)
    
    # %%
    result['RestaurantName'] = result['RestaurantName'].apply(lambda x: 'سطح شهر' if 'سطح شهر' in x else x)
    
    # %%
    def select_restaurant_type(x):
        if 'سطح شهر' in x:
            return 'khabgah'
        if 'خوابگاه' in x:
            return 'khabgah'
        return 'daneshgah'
    result['RestaurantType'] = result['RestaurantName'].apply(select_restaurant_type)
    
    # %%
    result['RestaurantName'].unique()
    
    # %%
    def remove_restaurant_type(x):
        for i in ['خوابگاه', 'دانشکده', 'پردیس', 'سلف']:
            x = x.replace(i, '')
        return x
    result['RestaurantName'] = result['RestaurantName'].apply(remove_restaurant_type)
    
    # %%
    def fix_campus_name(x):
        if 'کاسپین' in x:
            return 'کاسپین'
        elif 'فارابی' in x:
            return 'فارابی'
        elif 'ابوریحان' in x:
            return 'ابوریحان'
        elif 'کشاورزی' in x:
            return 'کشاورزی'
        elif 'فومن' in x:
            return 'فومن'
        return x
    result['RestaurantName'] = result['RestaurantName'].apply(fix_campus_name)
    
    # %%
    def fix_restaurant_name(x):
        for i in ['دکترا', 'کارکنان']:
            x = x.replace(i, '')
        for i in ['فنی1', 'فنی2']:
            x = x.replace(i, 'فنی ')
        x = x.replace('-', '')
        x = x.replace('فنیامیرآباد', 'فنی امیرآباد')
        x = x.strip()
        return x
    result['RestaurantName'] = result['RestaurantName'].apply(fix_restaurant_name)
    
    # %%
    result 
    
    # %%
    result[result['RestaurantName'] == result['RestaurantName'].unique()[0]]
    
    # %%
    for i in result['FoodName'].unique():
        print(i)
    
    # %%
    def total_nember(row):
        return row['ReceiveWithCard']+row['ReceiveWithCode'] + row['DontReceive']
    result['Reservation'] = result.apply(total_nember,axis=1)
    
    # %%
    result
    
    # %%
    def delete_definition(x):
        sep = '('
        stripped = x.split(sep, 1)[0]
        return stripped
    result['FoodName'] = result['FoodName'].apply(delete_definition)
    
    # %%
    result['FoodName'].nunique()
    
    # %%
    result.reset_index(drop=True)
    
    # %%
    result['FoodName'].unique()
    
    # %%
    result = result[result['FoodName'] != 'خرید از بوفه روزانه']
    result = result[result['FoodName'] != 'خرید از بوفه هفتگی ']
    result = result[result['FoodName'].str.contains('بسته ارزاق') == False]
    result = result.reset_index(drop = True)
    result['FoodName'].unique()
    
    # %%
    def select_name(x):
        if x == 'شوید باقلاباگوشت':
            return 'چلو شوید باقلاباگوشت'
        elif x == 'استانبولی با گوشت':
            return 'استانبولی پلو با گوشت'
        elif 'نودالیت' in x:
            return 'خوراک نودلیت'
        elif 'دمی' in x:
            return 'پلو' + x
        elif 'کتلت' in x:
            return 'خوراک کتلت'
        elif 'دلمه' in x:
            return 'خوراک دلمه'
        elif 'خوراك' in x:
            return x.replace('خوراك','خوراک')
        elif 'کنسرو تن ماهی' in x:
            return 'خوراک کنسرو تن ماهی'
        elif 'هواری مرغ' in x:
            return 'خوراک هواری مرغ'
        else:
            return x
    result['FoodName'] = result['FoodName'].apply(select_name)
    
    # %%
    def select_type(x):
        if 'چلو' in x:
            return "berenji"
        elif 'پلو' in x:
            return "berenji"
        elif 'خوراک' in  x:
            return "khorak"
        elif 'ساندویچ' in  x:
            return "khorak"
    
    # %%
    result['FoodType'] = result['FoodName'].apply(select_type)
    result['FoodType'].unique()
    
    # %%
    result
    
    # %%
    def delete__x000D(x):
        return x.replace("_x000D_\n","")
    result['Gender']=result['Gender'].apply(delete__x000D)
    
    # %%
    def convert_gender(x):
        if x == 'زن':
            return 'woman'
        return 'man'
    result['Gender'] = result['Gender'].apply(convert_gender)
    
    # %%
    def fix_meal(x):
        if 'صبحانه' in x:
            return 'breakfast'
        elif 'ناهار' in x:
            return 'lunch'
        elif 'شام' in x:
            return 'dinner'
        return x
    result['Meal'] = result['Name'].apply(fix_meal)
    result.drop('Name',axis=1,inplace=True)
    
    # %%
    result
    
    # %%
    grouped = result.groupby(['DateReserve','RestaurantName','RestaurantType','Meal','FoodName','FoodType','Gender'])
    result = grouped.agg({
        'ReceiveWithCard': 'sum',
        'ReceiveWithCode': 'sum',
        'DontReceive': 'sum',
        'Reservation': 'sum'
    }).reset_index()
    result
    
    # %%
    result = result[result['Reservation']> 3]
    result = result.reset_index(drop=True)
    
    # %%
    from persiantools.jdatetime import JalaliDate
    def convert_date_to_weekday(d):
        '''
        d: str, date in format 'yyyy/mm/dd'
        return: int, weekday of the date in range [0, 6]
        '''
        return JalaliDate(*map(int, d.split('/'))).weekday()
    convert_date_to_weekday(result['DateReserve'].iloc[0])
    result['DayOfWeek'] = result['DateReserve'].map(convert_date_to_weekday)
    
    # %%
    result
    
    # %%
    def str_to_date(d, gregorian:bool=False) -> JalaliDate:
        out = JalaliDate(*map(int, d.split('/')))
        if gregorian:
            return out.to_gregorian()
        return out
    
    date_map_gregorian = {i: str_to_date(i,True) for i in df['DateReserve'].unique()}
    result['DateReserveGregorian'] = result['DateReserve'].map(date_map_gregorian)
    
    date_map_jalali = {i: str_to_date(i) for i in df['DateReserve'].unique()}
    result['DateReserve'] = result['DateReserve'].map(date_map_jalali)
    
    result.info()
    
    # %%
    result
    
    # %%
    result['RestaurantName']
    
    # %%
    from bs4 import BeautifulSoup
    from holiday import Holiday, HolidayCalculator
    holidays = []
    for i in [1401,1402,1403]:
        with open(f"date/date_{i}.html", 'r') as file:
            file_content = file.read()
            soup = BeautifulSoup(file_content, 'html.parser')
            h = Holiday(soup)
            holidays_year = h.get_holidays(add_noroze=True, add_thursday=True)
            holidays.extend(holidays_year)
    holidays
    
    # %%
    result['DateReserve'].unique()
    
    # %%
    n = 2
    h_map = dict()
    for day in result['DateReserve'].unique():
        calculator = HolidayCalculator(holidays, day, have_weekend=False)
        h_map[day] = {
            'HolidayInWeekCount': calculator.get_holidays_this_week(True),
            'HolidayInPrevWeekCount': calculator.get_holidays_this_week(True,'Prev'),
            'HolidayInNextWeekCount': calculator.get_holidays_this_week(True,'Next'),
            # 'HolidayInWeek': calculator.get_holidays_this_week(),
            'NextHoliday': calculator.get_n_next_holiday(n),
            'PreviousHoliday': calculator.get_n_previous_holiday(n)
        }
    
    for i in h_map[result['DateReserve'][0]].keys():
        if i in ['PreviousHoliday', 'NextHoliday']:
            for j in range(n):
                result[f'{i}_{j+1}'] = result['DateReserve'].map(lambda d: h_map[d][i][j])
        else:
            result[i] = result['DateReserve'].map(lambda d: h_map[d][i])
    
    result = result[[col for col in result if col != 'DateReserveGregorian'] + ['DateReserveGregorian']]
    

    display(pd.DataFrame(h_map).T)
    display(result)
    return result

import os

result = list()
for file_name in os.listdir('./data'):
    if file_name.endswith('.xlsx'):
        tmp = run('./data/' + file_name)
        print(tmp['DateReserve'].iloc[0])
        result.append(tmp)
```
