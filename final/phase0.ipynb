```python
import pandas as pd
from jdatetime import datetime, date
from IPython import display
```

```python
df = pd.read_excel("data.xlsx")
```

```python
df['ReserveStatus'].unique()
```

```python
def convert_ReserveStatus(x):
    if(x == 'دریافت نشده' or x == 'ارسال نشده' or x == 'منقضی شده'):
        return 0
    return 1
```

```python
df['ReserveStatus'] = df['ReserveStatus'].apply(convert_ReserveStatus)
```

```python
df['ReserveStatus'].unique()
```

```python
df['Reception'].unique()
```

```python
def convert_Reception(x):
    if(x == 'کد فراموشی' or x == 'بارکد فراموشی'):
        return 1
    return 0
```

```python
df['Reception'] = df['Reception'].apply(convert_Reception)
```

```python
df['Reception'].unique()
```

```python
def apply_recive_with_card(row):
    if row['Reception'] == 0 and row['ReserveStatus'] == 1:
        return 1
    return 0
```

```python
df['reciev_with_card'] = df.apply(apply_recive_with_card,axis=1)
```

```python
def apply_reciev_with_code(row):
    if row['Reception'] == 1 and row['ReserveStatus'] == 1:
        return 1
    return 0
```

```python
df['reciev_with_code'] = df.apply(apply_reciev_with_code,axis=1)
```

```python
def apply_dont_reciev(row):
    if row['ReserveStatus'] == 0:
        return 1
    return 0
```

```python
df['dont_reciev'] = df.apply(apply_dont_reciev,axis=1)
```

```python
def select_name(x):
    if x == 'شوید باقلاباگوشت':
        return 'چلو شوید باقلاباگوشت'
    elif x == 'استانبولی با گوشت':
        return 'استانبولی پلو با گوشت'
    else:
        return x
result['FoodName'] = result['FoodName'].apply(select_name)
```

```python
def select_type(x):
    if 'چلو' in x:
        return "berenji"
    elif 'پلو' in x:
        return "berenji"
    elif 'خوراک' in  x:
        return "khorak"
    elif 'ساندویچ' in  x:
        return "khorak"
```

```python
result['type'] = result['FoodName'].apply(select_type)
result['type'].unique()
```

```python
from persiantools.jdatetime import JalaliDate
def convert_date_to_weekday(d):
    '''
    d: str, date in format 'yyyy/mm/dd'
    return: int, weekday of the date in range [0, 6]
    '''
    return JalaliDate(*map(int, d.split('/'))).weekday()
convert_date_to_weekday(result['DateReserve'].iloc[0])
result['DayOfWeek'] = result['DateReserve'].map(convert_date_to_weekday)
```

```python
result
```

```python
def str_to_date(d, gregorian:bool=False) -> JalaliDate:
    out = JalaliDate(*map(int, d.split('/')))
    if gregorian:
        return out.to_gregorian()
    return out
result['DateReserve'] = df['DateReserve'].map(str_to_date)
result['DateReserveGregorian'] = df['DateReserve'].map(lambda d: str_to_date(d, True)).astype('datetime64[ns]')

result.info()
```

```python
from bs4 import BeautifulSoup
from holiday import Holiday, HolidayCalculator
holidays = []
for i in [1401,1402,1403]:
    with open(f"date/date_{i}.html", 'r') as file:
        file_content = file.read()
        soup = BeautifulSoup(file_content, 'html.parser')
        h = Holiday(soup)
        holidays_year = h.get_holidays(add_noroze=True, add_thursday=True)
        holidays.extend(holidays_year)
holidays
```

```python
h_map = dict()
n = 5
for day in result['DateReserve'].unique():
    calculator = HolidayCalculator(holidays, day, have_weekend=False)
    h_map[day] = {
        'HolidayInWeekCount': calculator.get_holidays_this_week(True),
        'HolidayInWeek': calculator.get_holidays_this_week(),
        'NextHoliday': calculator.get_n_next_holiday(n),
        'PreviousHoliday': calculator.get_n_previous_holiday(n)
    }

result['HolidayInWeekCount'] = result['DateReserve'].map(lambda d: h_map[d]['HolidayInWeekCount'])

for i in ['PreviousHoliday', 'NextHoliday']:
    for j in range(5):
        result[f'{i}_{j+1}'] = result['DateReserve'].map(lambda d: h_map[d][i][j])

display(pd.DataFrame(h_map).T)
display(result)
```
