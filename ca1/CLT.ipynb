# CLT Part

**part 1:**

in the first section, we import our necessary libraries.

numpy is used for arithmetic operations

 matplotlib.pyplot for plotting graphs

  and scipy.stats for working with probability distributions.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, bernoulli, uniform, probplot, expon
```

**part 2:**

in this section, we create our distributions. here we used normal, Bernoulli and uniform distributions and determined the parameters.

```python

exponential_params = {"scale": 1}
bernoulli_params = {"p": 0.5}
uniform_params = {"loc": 0, "scale": 1}


distributions = {
    "exponential": expon(**exponential_params),
    "Bernoulli": bernoulli(**bernoulli_params),
    "Uniform": uniform(**uniform_params)
}
```

part 3:


in this section, we define our sample sizes

```python
# Define the sample sizes
sample_sizes = [1, 10, 100,500,1000]

# Define the number of samples
num_samples = 1000
```

```python


# For each distribution
for dist_name, dist in distributions.items():
    print(f"\n{dist_name} Distribution:")

    # For each sample size
    for sample_size in sample_sizes:
        sample_means = []

        # Generate the samples and calculate their means
        for _ in range(num_samples):
            sample = dist.rvs(size=sample_size)
            sample_means.append(np.mean(sample))

         # Create a figure with two subplots
        fig, axs = plt.subplots(1, 2, figsize=(10, 5))

        # Plot the histogram of the sample means in the first subplot
        axs[0].hist(sample_means, bins=30, density=True, alpha=0.6, color='g')

        # Overlay the expected normal distribution
        mu, std = norm.fit(sample_means)
        xmin, xmax = axs[0].get_xlim()
        x = np.linspace(xmin, xmax, 100)
        p = norm.pdf(x, mu, std)
        axs[0].plot(x, p, 'k', linewidth=2)
        title = f"Fit results: mu = {mu:.2f},  std = {std:.2f}"
        axs[0].set_title(title)

        # Plot the Q-Q plot for the sample means in the second subplot
        probplot(sample_means, dist="norm", plot=axs[1])
        axs[1].set_title(f"Q-Q plot for sample size {sample_size}")

        # Show the figure
        plt.tight_layout()
        plt.show()
```

According to the results of the code execution, several important points can be observed:

**The distribution of the sample mean approaches a normal distribution:** As the sample size increases, the distribution of the sample mean approaches a normal distribution. This is one of the main results of the central limit law.


**Increasing accuracy with increasing sample size:** As the sample size increases, the distribution of the mean of the samples becomes closer to the population mean. This suggests that as the sample size increases, our estimate of the population mean becomes more accurate.


**Reducing the standard deviation by increasing the sample size:** By increasing the sample size, the standard deviation of the average distribution of the samples decreases. This indicates that as the sample size increases, the distribution of the sample mean becomes more concentrated, which means that our estimate of the population mean is more reliable.


**Independence from the initial distribution:** the central limit law holds for any probability distribution, whether this distribution is normal or not. In this experiment, we observed this for three different distributions: normal, Bernoulli, and uniform.


These results show how the central limit law can help us in practice to obtain more accurate estimates of population parameters based on sample statistics.
